```python
import numpy as np
import os, sys, time, subprocess, h5py, argparse, logging, pickle, random
from os.path import join as oj
import pandas as pd
import matplotlib
matplotlib.use('agg')
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# get filenames
results_dir = '/Users/chandan/drive/research/vision/scat/scat_results/cifar100'
log_files = {}
for dirname in os.listdir(oj(results_dir, 'checkpoint')):
    if '_n2' in dirname:
        log_files[dirname[:-3]] = oj(results_dir, 'checkpoint', dirname, 'log.txt')
        
for key, val in log_files.items():
    print(key, val)
```

```python
out_dir = '/Users/chandan/drive/research/vision/scat/scat_results/figs/validation_curves'
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
for key, val in log_files.items():
    print(key, val)
    x = np.loadtxt(val, skiprows=1)
    print(x.shape)
    plt.plot(x[:, -1], label='Validation accuracy')
    plt.xlabel('Iterations')
    plt.ylabel('Cifar100 validation accuracy')
    plt.ylim((0, 45))
    plt.savefig(oj(out_dir, key + '.png'))
    plt.show()
```
